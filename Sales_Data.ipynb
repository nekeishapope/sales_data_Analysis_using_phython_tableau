```python
data = pd.read_csv("sales_data_sample.csv", sep=",", encoding='Latin-1')
df = pd.DataFrame(data)
canada = df['COUNTRY'].value_counts()['Canada']
usa = df['COUNTRY'].value_counts()['USA']
uk = df['COUNTRY'].value_counts()['UK']


def country_bar_chart(country, values):
    plt.bar(country, values, color='blue')
    plt.title('Total Products Sold in UK CANADA and USA')
    plt.show()

country = ['Canada', 'USA', 'UK']
values = [canada, usa, uk]
country_bar_chart(country, values)
```
